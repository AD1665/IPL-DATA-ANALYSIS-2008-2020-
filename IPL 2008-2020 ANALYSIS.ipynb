# MOST WINS IN IPL

```python
print("Teams which has most wins are")
match_data['winner'].value_counts()
```

```python
plt.figure(figsize=(10,10))
data=match_data['winner'].value_counts()
sns.barplot(y=data,x=data.index)
plt.title("Most wins in IPL")
plt.xticks(rotation=90)
plt.ylabel("wins")
plt.show()
```

# MOST WINS IN ELIMINATOR

```python
sns.countplot(match_data['winner'][match_data['eliminator']=='Y'],data=match_data)
plt.title("Most wins in Eliminator",fontsize=17)
plt.xticks(rotation=90)
plt.show()
```

```python
match_data['Season'] = pd.DatetimeIndex(match_data['date']).year
```

# NUMBER OF MATCHES PER SEASON

```python
match_per_season = match_data.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season
```

```python
sns.countplot(match_data['Season'])
plt.xticks(rotation=45,fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("Season",fontsize=20)
plt.ylabel("count",fontsize=20)
plt.title("Number of matches played each season",fontsize=20,fontweight="bold")
plt.show()
```

```python
season_data=match_data[['id','Season']].merge(ball_data, left_on = 'id', right_on = 'id', how = 'left').drop('id', axis = 1)
season_data.head()
```

# TOTAL RUNS MADE IN EACH SEASON

```python
season = season_data.groupby(["Season"])['total_runs'].sum().reset_index()
season
```

```python
p=season.set_index('Season')
ax = plt.axes()
ax.set(facecolor = "grey")
sns.lineplot(data=p,palette="magma") 
plt.title('Total runs in each season',fontsize=20,fontweight="bold")
plt.show()
```

# RUNS SCORED PER MATCH IN EACH SEASON

```python
runs_per_season=pd.concat([match_per_season,season.iloc[:,1]],axis=1)
runs_per_season['Runs scored per match']=runs_per_season['total_runs']/runs_per_season['matches']
runs_per_season.set_index('Season',inplace=True)
runs_per_season
```

# ORANGE CAP HOLDER EACH SEASON

```python
data = season_data.groupby(['Season','batsman'])['batsman_runs'].sum().groupby('Season').max()
temp_df=pd.DataFrame(season_data.groupby(['Season','batsman'])['batsman_runs'].sum())
print("{0:10}{1:20}{2:30}".format("Season","Player","Runs"))
for season,run in data.items():
    player = temp_df.loc[season][temp_df.loc[season]['batsman_runs'] == run].index[0]
    print(season,'\t ',player,'\t\t',run)
```

# MOST 6 by a Batsman

```python
temp_df = ball_data.groupby('batsman')['batsman_runs'].agg(lambda x:(x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).head(10).reset_index(drop=True)
temp_df
sns.barplot(x=temp_df['batsman'],y=temp_df['batsman_runs'],data=temp_df)

plt.title("Most sixes")
plt.xticks(rotation=90)
plt.xlabel("Batsman")
plt.ylabel("Number of 6's")
plt.show()
```
